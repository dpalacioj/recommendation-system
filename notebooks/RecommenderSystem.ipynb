```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter +=1

    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fechtch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fechtch_director)
```

```python
movies.head(1)
```

Now, we have all the columns in the correcto format to work with it.

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

Some words are repeated also in `cast` as in `crew`. It could generate a confuse escenario.
For that reason, we have to use strategies as:

* Sam Worthington --> SamWorthington

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ", ""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

Now we have to concatenate the columns to generate a chain of text for use later in a recommender system

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head(2)
```

```python
# Now, we have to convert it into paragraph

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))

new_df.head(2)
```

```python
# Now we have to convert the text into lower case

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head(2)
```

You could find similar words in the colum 'tags'. <br/>
For instance: action and actions; love, loved, loving. <br/>
For that reason we need fix the paragraph:

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# For reduce the words to its root
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
print(vector.shape)
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df.head()
```

```python
new_df[new_df['title'] == 'Spider-Man']
```

```python
similarity[159]
```

```python
distances = sorted(list(enumerate(similarity[159])), reverse=True, key = lambda x: x[1])
distances
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('The Dark Knight')
```
